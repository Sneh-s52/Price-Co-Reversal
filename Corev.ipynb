```python
import yfinance as yf
import pandas as pd
import pandas as pd
```

```python
stocks = ['MSFT','AMZN','AAPL','NFLX','META','GOOGL','TSLA','NVDA','AVGO','ORCL','ADBE','CRM','CSCO','^GSPC']
stocks_data = yf.download(stocks, start='2023-01-01',interval='1wk')
stocks_data = stocks_data['Adj Close']
stocks_data.head()
```

```python
returns_data = stocks_data[stocks].pct_change().drop('2023-01-01')
SP_500_ret = returns_data['^GSPC']
returns_data = returns_data.drop(columns='^GSPC')
print(SP_500_ret.head())
print(returns_data.head())
abnormal_returns = returns_data.sub(SP_500_ret, axis=0)
abnormal_returns.head()
```

```python
abnormal_returns.corr()
```

```python
# Calculate past cumulative returns for each stock
cumulative_returns = (1 + returns_data).cumprod() - 1
cumulative_returns.iloc[-1]
# Sort stocks based on the last cumulative returns
sorted_tickers = (cumulative_returns.iloc[-1]).sort_values(ascending=False).index

# Print the sorted tickers
print(sorted_tickers)
```

```python
# Calculate percentiles for each stock
percentiles = pd.qcut(cumulative_returns.iloc[-1], q=10, labels=False) + 1

# Create a DataFrame to store tickers and percentiles
# result_df = pd.DataFrame({'Ticker': sorted_tickers, 'Decile': percentiles})

# Print the resulting DataFrame
print(percentiles)
```

```python
sum = 0

for i in stocks:
    target_stock = i  # Replace with the desired stock symbol

    # Extract the excess returns of the target stock
    target_stock_returns = abnormal_returns[target_stock]

    # Calculate equal-weighted average of excess returns for other stocks
    other_stocks_returns = abnormal_returns.drop(columns=target_stock)
    equal_weighted_returns = other_stocks_returns.mean(axis=1)

    # Calculate the correlation between the target stock's excess return and the equal-weighted returns of other stocks
    correlation = target_stock_returns.corr(equal_weighted_returns)
    sum = sum + correlation

    # Print the correlation value
    print(f"Correlation between {target_stock} and other stocks (weighted equally): {correlation}")
 
CoRev = sum/len(stocks)
print(f"The CoRev obtained from above stocks is ",CoRev)
```
